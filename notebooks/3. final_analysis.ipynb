```python
# Creating ranges of Train Weight

def weight_train(x):
    if x <= 1000:
        return '<1000'
    if 1000 < x and x <= 1250:
        return '1000-1250'
    if 1250 < x and x <= 1500:
        return '1250-1500'
    if 1500 < x and x <= 1750:
        return '1500-1750'
    return '>1750'

def length_train(x):
    if x <= 500:
        return '<500'
    if 500 < x and x <= 550:
        return '500-550'
    if 550 < x and x <= 650:
        return '550-650'
    return '>650'

def wagons_count(x):
    if x <= 15:
        return '<15'
    if 15 < x and x <= 25:
        return '15-25'    
    return '>25'

def weight_per_wagon(x):
    if x <= 40:
        return '<40'
    if 40 < x and x <= 60:
        return '40-60'
    if 60 < x and x <= 80:
        return '60-80'
    return '>80'

def distance_km(x):
    if x <= 200:
        return '<200'
    if 200 < x and x <= 500:
        return '200-500'
    return '>500'

df_predictors['weight_bin'] = df_predictors['Train_Weight'].apply(lambda x: weight_train(x))
df_predictors['length_bin'] = df_predictors['Train_Length'].apply(lambda x: length_train(x))
df_predictors['wagon_count_bin'] = df_predictors['wagon_count'].apply(lambda x: wagons_count(x))
df_predictors['weight_per_wagon_bin'] = df_predictors['weight_wagon'].apply(lambda x: weight_per_wagon(x))
df_predictors['distance_bin'] = df_predictors['Train_Distance_KM'].apply(lambda x: distance_km(x))
```

```python
# Train Weight
df_predictors['weight_bin'] = pd.Categorical(df_predictors['weight_bin'],
                                                categories=['<1000', '1000-1250', '1250-1500', '1500-1750', '>1750'],
                                                ordered=True)


# Analysing statistical correlations between different parameters
plt.figure(figsize=(80,40))

dfP = pd.crosstab(index=df_predictors['weight_bin'], columns=df_predictors['arrived'])
dfP.plot(kind='bar')

plt.xticks(rotation=0)
plt.xlabel('Train Weight [t]', fontsize='large')
plt.tick_params(labelsize=10)
#plt.title("Choice vs. travel time")
   
    
L=plt.legend(loc='best', fontsize=10)
L.get_texts()[0].set_text('Non-delayed trains')
L.get_texts()[1].set_text('Delayed trains')


Ma = dfP.sum().sum()             #Ma = 8658 en caso de que quieran ver la comparacion con EL TOTAL DE TODA LA MUESTRA
plt.yticks(ticks = np.linspace(0,Ma, 21), labels=["{:.0f}%".format(i*100) for i in np.linspace(0, 1, 21)])
plt.margins(0,0.3)

plt.savefig("train_weight.png", dpi=400,  bbox_inches='tight')
plt.show();
```
